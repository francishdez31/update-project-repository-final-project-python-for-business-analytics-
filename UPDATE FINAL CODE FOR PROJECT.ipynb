```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'C:\\Users\\franc\\Documents\\josh\\Dataset update.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset and column names
print("First few rows of the dataset:")
print(data.head())
print("\nColumn names in the dataset:")
print(data.columns)

# Setting the style for the plots
sns.set(style="whitegrid")

# Define performance metrics
data['Revenue'] = data['Sales']  # Assuming Sales column represents revenue

# Check for the presence of a 'Cost' column
if 'Cost' in data.columns:
    data['Profit'] = data['Sales'] - data['Cost']
    data['Profit_Ratio'] = data['Profit'] / data['Sales']
else:
    print("Warning: 'Cost' column not found. Profit and Profit Ratio calculations will be skipped.")

# 1. Total Revenue by Product
product_revenue = data.groupby(['Product_Card_Id', 'Category_Name'])['Revenue'].sum().reset_index()
product_revenue = product_revenue.sort_values(by='Revenue', ascending=False)
plt.figure(figsize=(15, 10))
sns.barplot(x='Revenue', y='Category_Name', data=product_revenue, palette="viridis")
plt.xlabel('Total Revenue')
plt.ylabel('Product Category')
plt.title('Total Revenue by Product')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 2. Total Revenue by Region
region_revenue = data.groupby('Order_Region')['Revenue'].sum().reset_index()
region_revenue = region_revenue.sort_values(by='Revenue', ascending=False)
plt.figure(figsize=(15, 10))
sns.barplot(x='Order_Region', y='Revenue', data=region_revenue, palette="plasma")
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.title('Total Revenue by Region')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 3. Sales Trends Over Time (if Date column exists)
if 'Date' in data.columns:
    data['Date'] = pd.to_datetime(data['Date'])
    sales_trends = data.groupby(data['Date'].dt.to_period('M'))['Revenue'].sum().reset_index()
    sales_trends['Date'] = sales_trends['Date'].dt.to_timestamp()
    plt.figure(figsize=(15, 10))
    sns.lineplot(x='Date', y='Revenue', data=sales_trends, marker='o', color='b')
    plt.xlabel('Date')
    plt.ylabel('Total Revenue')
    plt.title('Sales Trends Over Time')
    plt.grid(True)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("\nError: The dataset does not contain a 'Date' column. Skipping sales trends analysis.")

# 4. Customer Segmentation
customer_revenue = data.groupby('Order_Customer_Id')['Revenue'].sum().reset_index()
customer_revenue = customer_revenue.sort_values(by='Revenue', ascending=False)
top_customers = customer_revenue.head(10)
plt.figure(figsize=(15, 10))
sns.barplot(x='Order_Customer_Id', y='Revenue', data=top_customers, palette="magma")
plt.xlabel('Customer ID')
plt.ylabel('Total Revenue')
plt.title('Top 10 Customers by Revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 5. Product Performance Analysis
product_performance = data.groupby(['Category_Name'])['Revenue'].mean().reset_index()
plt.figure(figsize=(15, 10))
sns.barplot(x='Revenue', y='Category_Name', data=product_performance, palette="coolwarm")
plt.xlabel('Average Revenue')
plt.ylabel('Product Category')
plt.title('Average Revenue by Product Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 6. Monthly Revenue Trends (if Date column exists)
if 'Date' in data.columns:
    data['Month'] = data['Date'].dt.to_period('M')
    monthly_revenue = data.groupby('Month')['Revenue'].sum().reset_index()
    plt.figure(figsize=(15, 10))
    sns.lineplot(x='Month', y='Revenue', data=monthly_revenue, marker='o', color='g')
    plt.xlabel('Month')
    plt.ylabel('Total Revenue')
    plt.title('Monthly Revenue Trends')
    plt.grid(True)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# 7. Average Sales Price by Product
average_price = data.groupby('Product_Card_Id')['Order_Item_Product_Price'].mean().reset_index()
plt.figure(figsize=(15, 10))
sns.barplot(x='Product_Card_Id', y='Order_Item_Product_Price', data=average_price, palette="cubehelix")
plt.xlabel('Product ID')
plt.ylabel('Average Sales Price')
plt.title('Average Sales Price by Product')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# 8. Revenue Distribution by Region
plt.figure(figsize=(15, 10))
sns.boxplot(x='Order_Region', y='Revenue', data=data, palette="Set2")
plt.xlabel('Region')
plt.ylabel('Revenue')
plt.title('Revenue Distribution by Region')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 9. Category-wise Revenue Distribution
plt.figure(figsize=(15, 10))
sns.boxplot(x='Category_Name', y='Revenue', data=data, palette="Set3")
plt.xlabel('Product Category')
plt.ylabel('Revenue')
plt.title('Revenue Distribution by Product Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 10. Correlation Heatmap (only for numeric columns)
numeric_data = data.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(15, 10))
correlation = numeric_data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Regional Comparison and Opportunity Identification

# Assuming 'Order_Region', 'Revenue', 'Sales', 'Product_Card_Id', 'Order_Customer_Id' columns exist

if 'Profit' in data.columns:
    # Total Profit by Region
    region_profit = data.groupby('Order_Region')['Profit'].sum().reset_index()
    region_profit = region_profit.sort_values(by='Profit', ascending=False)
    plt.figure(figsize=(15, 10))
    sns.barplot(x='Order_Region', y='Profit', data=region_profit, palette="coolwarm")
    plt.xlabel('Region')
    plt.ylabel('Total Profit')
    plt.title('Total Profit by Region')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Identifying regions with negative profit
    negative_profit_regions = region_profit[region_profit['Profit'] < 0]
    print("Regions with negative profit:")
    print(negative_profit_regions)
else:
    print("Profit analysis skipped due to missing 'Cost' column.")

# Top 3 regions with the highest profit
top_regions = region_revenue.head(3)['Order_Region'].values

# Compare top 3 regions
for region in top_regions:
    region_data = data[data['Order_Region'] == region]
    plt.figure(figsize=(15, 10))
    sns.barplot(x='Category_Name', y='Revenue', data=region_data.groupby('Category_Name')['Revenue'].sum().reset_index(), palette="viridis")
    plt.xlabel('Product Category')
    plt.ylabel('Total Revenue')
    plt.title(f'Total Revenue by Product Category in {region}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Customer analysis in the region
    customer_revenue = region_data.groupby('Order_Customer_Id')['Revenue'].sum().reset_index()
    top_customers = customer_revenue.sort_values(by='Revenue', ascending=False).head(10)
    plt.figure(figsize=(15, 10))
    sns.barplot(x='Order_Customer_Id', y='Revenue', data=top_customers, palette="magma")
    plt.xlabel('Customer ID')
    plt.ylabel('Total Revenue')
    plt.title(f'Top 10 Customers by Revenue in {region}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
